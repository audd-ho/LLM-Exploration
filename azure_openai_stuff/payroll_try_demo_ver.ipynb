### Azure OpenAI Client Setup

```python
import os
from openai import AzureOpenAI

endpoint = "https://openai-iras-payroll-assurance.openai.azure.com/" # Azure OpenAI endpoint
# key = os.getenv("AZURE_OPENAI_API_KEY")
key = "xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"

client = AzureOpenAI(
    api_key= key,  
    api_version="2024-02-01",
    azure_endpoint = endpoint
)
```

### ChatGPT Example

```python
completion = client.chat.completions.create(
    model="gpt-35-turbo", # model name (may not be same as public openai), only deployed models are available
    messages=[
        {
            "role": "system",
            "content": "you are a HR chat bot that always say 'Good day! everyone has 18 days leave annually'",
        },
        {
            "role": "user",
            "content": "how many days of leave do I have?"
        }
    ]  
)
```

```python
response_msg = completion.choices[0].message.content
print(response_msg)
```

### Text Embedding Example

```python
embedding = client.embeddings.create(
    model="text-embedding-3-large",
    input = response_msg
) 
```
